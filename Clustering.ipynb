#**Clustering**


### **1. What is Unsupervised Learning in Machine Learning?**

**Unsupervised learning** is a type of machine learning where the model is **not given labeled data**. The goal is to **find patterns or structure** in the data.

#### üîç Main tasks:

* **Clustering** (e.g., grouping customers by behavior)
* **Dimensionality Reduction** (e.g., PCA)
* **Anomaly Detection**

 There are **no target labels** ‚Äî the model tries to learn the data's **hidden structure**.

---

### **2. How Does K-Means Clustering Algorithm Work?**

**K-Means** partitions data into **K clusters** such that **intra-cluster distance is minimized**.

####  Steps:

1. Choose number of clusters $K$
2. Randomly initialize $K$ cluster centroids
3. Assign each point to the nearest centroid (E-step)
4. Update centroids as the mean of all points in that cluster (M-step)
5. Repeat steps 3‚Äì4 until centroids stop changing (convergence)

 Goal: Minimize **inertia** (sum of squared distances to centroids)

---

### **3. Explain the Concept of a Dendrogram in Hierarchical Clustering**

A **dendrogram** is a **tree-like diagram** that shows how data points are merged (or split) in **hierarchical clustering**.

#### üîπ Use:

* Visualize the **nested grouping** of clusters
* Decide **how many clusters** to select by "cutting" the tree at a specific height

 It shows the **distance at which clusters were merged**, helping to determine cluster structure.

---

### **4. Main Difference Between K-Means and Hierarchical Clustering**

| Feature           | **K-Means**            | **Hierarchical**                |
| ----------------- | ---------------------- | ------------------------------- |
| **Approach**      | Partitional            | Hierarchical                    |
| **Output**        | Fixed K clusters       | Tree structure (dendrogram)     |
| **Scalability**   | Fast on large datasets | Slow on large datasets          |
| **Cluster Shape** | Spherical              | Can be non-spherical            |
| **Requires K?**   | Yes                    | No (can decide from dendrogram) |

---

### **5. Advantages of DBSCAN Over K-Means**

| Feature                                          | **DBSCAN** |
| ------------------------------------------------ | ---------- |
| No need to specify **K**                         |            |
| Detects **arbitrary-shaped** clusters            |            |
| Identifies **noise/outliers**                    |            |
| Handles **non-linear boundaries**                |            |
| More robust to **scale and density differences** |            |

 Especially useful for **spatial, noisy, and real-world data**.

---

### **6. When Would You Use Silhouette Score in Clustering?**

Use **Silhouette Score** to **evaluate the quality** of clustering.

####  Formula:

$$
\text{Silhouette Score} = \frac{b - a}{\max(a, b)}
$$

Where:

* $a$: average distance to points in the same cluster
* $b$: average distance to points in the nearest cluster

#### üìà Interpretation:

* +1: well-clustered
* 0: borderline
* -1: wrongly clustered

 Use it to **choose optimal K** in K-Means or **compare clustering results**.

---

### **7. Limitations of Hierarchical Clustering**

* **Not scalable** (complexity = $O(n^2)$)
* **Irreversible merges/splits** (can‚Äôt fix early mistakes)
* Sensitive to **outliers**
* Requires a **distance metric** (choice affects results)
* **No optimization objective** like inertia or log-likelihood

---

### **8. Why is Feature Scaling Important in Clustering (e.g., K-Means)?**

Because clustering uses **distance-based measures** (e.g., Euclidean distance):

* Features with **larger ranges** dominate the distance
* Leads to **biased clustering**

 Apply **standardization (Z-score)** or **min-max scaling** before clustering

---

### **9. How Does DBSCAN Identify Noise Points?**

DBSCAN uses two parameters:

* `eps`: neighborhood radius
* `min_samples`: minimum points in a dense region

####  Point Classification:

* **Core point**: ‚â• `min_samples` in `eps` radius
* **Border point**: < `min_samples` but near core point
* **Noise point (outlier)**: not close to any core point

 Noise points are **unclustered**, useful for **anomaly detection**

---

### **10. Define Inertia in the Context of K-Means**

**Inertia** is the **sum of squared distances** between each point and its cluster centroid.

#### Formula:

$$
\text{Inertia} = \sum_{i=1}^{k} \sum_{x_j \in C_i} ||x_j - \mu_i||^2
$$

Where:

* $C_i$: cluster
* $\mu_i$: cluster centroid

 Lower inertia = **better clustering**, but adding more clusters always reduces inertia ‚Üí not ideal for model selection by itself.
---

### **11. What is the Elbow Method in K-Means Clustering?**

The **Elbow Method** helps determine the **optimal number of clusters (K)** in K-Means.

#### üîç How it works:

1. Run K-Means for a range of K values (e.g., 1 to 10).
2. Plot **inertia** (within-cluster sum of squares) vs. **K**.
3. Look for an **"elbow" point** ‚Äî the point after which adding more clusters **gives diminishing returns**.

‚úÖ The elbow is where the model balances **complexity and performance**.

---

### **12. Describe the Concept of "Density" in DBSCAN**

In DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

* **Density** = number of points within a given radius (`eps`)
* If a point has **‚â• min\_samples** in its `eps`-neighborhood ‚Üí it‚Äôs a **core point**

‚úÖ DBSCAN clusters are **dense regions of points** separated by low-density areas (noise).

---

### **13. Can Hierarchical Clustering Be Used on Categorical Data?**

Yes ‚Äî but **standard hierarchical clustering** (based on Euclidean distance) doesn‚Äôt handle categorical data directly.

#### ‚úÖ Solutions:

* Use **distance metrics** like:

  * **Hamming distance**
  * **Gower distance** (mixed-type data)
* Use **One-Hot Encoding** if ordinal relationship isn‚Äôt important
* Consider **hierarchical clustering for categorical attributes** (e.g., using similarity matrices)

---

### **14. What Does a Negative Silhouette Score Indicate?**

A **negative Silhouette Score** means:

* A point is **closer to a different cluster** than to its own
* Likely **misclassified** or poorly clustered

üìâ Indicates:

* **Incorrect number of clusters**
* Overlapping clusters
* Poor separation

---

### **15. Explain the Term "Linkage Criteria" in Hierarchical Clustering**

**Linkage criteria** determine **how the distance between clusters is calculated** when merging.

#### Common types:

| Linkage Type | Definition                                                          |
| ------------ | ------------------------------------------------------------------- |
| **Single**   | Minimum distance between two points (nearest neighbor)              |
| **Complete** | Maximum distance between points (farthest neighbor)                 |
| **Average**  | Average of all pairwise distances                                   |
| **Ward**     | Minimize variance within clusters (uses squared Euclidean distance) |

‚úÖ Choice affects **shape and sensitivity** of clusters.

---

### **16. Why Might K-Means Perform Poorly on Varying Cluster Sizes or Densities?**

Because K-Means:

* Assumes **spherical, equal-sized, equally dense clusters**
* Uses **Euclidean distance** ‚Üí sensitive to **scale and spread**
* Assigns points to **nearest centroid**, not densest region

üìå In such cases:

* K-Means **misclassifies** border points
* May **split larger clusters** or **merge smaller ones incorrectly**

---

### **17. Core Parameters in DBSCAN and Their Influence**

| Parameter     | Description                         | Influence                                 |
| ------------- | ----------------------------------- | ----------------------------------------- |
| `eps`         | Radius around a point               | Affects size of neighborhood              |
| `min_samples` | Minimum points to form dense region | Affects sensitivity to noise and outliers |

‚úÖ Large `eps` ‚Üí fewer, larger clusters
‚úÖ Small `min_samples` ‚Üí more clusters, more sensitivity to noise

---

### **18. How Does K-Means++ Improve Upon Standard K-Means Initialization?**

**K-Means++** improves **centroid initialization**, reducing the chance of poor convergence.

#### Steps:

1. Randomly pick the first centroid
2. For the rest, pick points **with probability proportional to their squared distance** from the nearest centroid

‚úÖ Leads to:

* **Faster convergence**
* **Better clustering**
* **Lower inertia**

---

### **19. What Is Agglomerative Clustering?**

**Agglomerative Clustering** is a **bottom-up hierarchical method**:

1. Start with each point as its **own cluster**
2. Iteratively **merge the closest clusters** based on **linkage criteria**
3. Continue until **all points are merged into one cluster (or until cutoff)**

‚úÖ Most common type of hierarchical clustering.

---

### **20. Why Is Silhouette Score Better Than Inertia for Model Evaluation?**

| Metric               | Why Silhouette Score is better                                              |
| -------------------- | --------------------------------------------------------------------------- |
| **Inertia**          | Always decreases as K increases ‚Üí **can‚Äôt detect optimal K**                |
| **Silhouette Score** | Measures **cohesion (within-cluster)** vs. **separation (between-cluster)** |

Silhouette Score gives a **balance** between compactness and separation
Works **regardless of cluster number**, and helps detect **bad clustering**

```python
#21 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Visualize the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')  # points colored by cluster
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centers')  # plot centers
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q. 22  Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10 predicted labels

from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data  # Using all four features

# Step 2: Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
labels = agg_clustering.fit_predict(X)

# Step 3: Display the first 10 predicted cluster labels
print("First 10 predicted labels:")
print(labels[:10])
```

```python
#Q. 23 Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Step 1: Generate synthetic moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Step 2: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)  # You can adjust eps and min_samples
labels = dbscan.fit_predict(X)

# Step 3: Plot the clusters and highlight outliers
plt.figure(figsize=(8, 6))

# Core and border points
plt.scatter(X[labels != -1][:, 0], X[labels != -1][:, 1],
            c=labels[labels != -1], cmap='viridis', s=50, label='Clusters')

# Outliers (noise points labeled as -1)
plt.scatter(X[labels == -1][:, 0], X[labels == -1][:, 1],
            c='red', s=50, marker='x', label='Outliers')

plt.title("DBSCAN Clustering with make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q. 26 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 3: Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.title('Silhouette Scores for k = 2 to 5')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(axis='y')
plt.show()
```

```python
#Q.27  Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Step 1: Generate overlapping synthetic data
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=1.5, random_state=42)

# Step 2: Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Step 3: Create meshgrid for decision boundaries
h = 0.02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries and data
plt.figure(figsize=(8, 6))
cmap_background = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_points = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.contourf(xx, yy, Z, cmap=cmap_background, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap_points, s=50, edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q. 30 Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding noise)

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Load the Wine dataset
data = load_wine()
X = data.data

# Step 2: Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)  # eps may need tuning
labels = dbscan.fit_predict(X_scaled)

# Step 4: Count the number of clusters (excluding noise = label -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Output
print("Cluster labels:", np.unique(labels))
print("Number of clusters (excluding noise):", n_clusters)
```
